```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
```

```python
data = pd.read_csv('FuelConsumption.csv')
data.head(5)
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
df = data[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
df.head(9)
```

```python
df.hist()
```

```python
plt.scatter(df.ENGINESIZE,df.CO2EMISSIONS,color='blue')
plt.xlabel('ENGINESIZE')
plt.ylabel('Emission')
plt.show()
     
```

```python
plt.scatter(df.FUELCONSUMPTION_COMB	,df.CO2EMISSIONS,color='blue')
plt.xlabel('FUELCONSUMPTION_COMB	')
plt.ylabel('Emission')
plt.show()
```

```python
ttd = np.random.rand(len(data))<0.8
train = df[ttd]
test = df[~ttd]
```

```python
plt.scatter(train.ENGINESIZE,train.CO2EMISSIONS,color='blue')
plt.xlabel('ENGINESIZE')
plt.ylabel('Emission')
plt.show()
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
model.fit(train_x,train_y)

#the coefficients

print('Coefficients ',model.coef_)
print('Intercept',model.intercept_)
```

```python
plt.scatter(train.ENGINESIZE,train.CO2EMISSIONS,color='blue')
plt.plot(train_x,model.coef_[0][0]*train_x+model.intercept_[0],'-r')
plt.xlabel('ENGINESIZE')
plt.ylabel('Emission')
plt.show()
```

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
prediction = model.predict(test_x)

# errors

print('Mean absolute error : %.2f' % np.mean(np.absolute(prediction-test_y)))
print('Residual sum of square(MSE) : %.2f' % np.mean((prediction-test_y)**2))
print('R2-score : %.2f' % r2_score(prediction,test_y))
```
